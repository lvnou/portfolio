## Distibution of risk

Risk classes are defined as follows:
* 0 - Liquid, availible without risk
* 1 - Low-risk bonds (government), self-owned real estate
* 2 - Higher-risk bonds, mutual funds with stocks + bonds
* 3 - Mutual funds in stocks, ETF, Index funds
* 4 - Individual stock
* 5 - Derivates etc.    

```python
rp = my_port.collect_risk_class()
rp = process_table(rp)
plt.pie(rp.value, labels=rp.risk_class)
plt.show()
```
